### Задача3

```python
test_data_numerical = ["bedrooms","bathrooms","sqft_living","sqft_lot","floors","sqft_above","sqft_basement","yr_built","yr_renovated"]
test_data_categorical = ["waterfront","view","condition","grade"]
```

```python
plt.figure(figsize=(10,5))
for i in test_data_numerical:
    sns.histplot(test_data[i],bins=100)
    plt.title(f'Метрика {i}')
    plt.xlabel(f"{i}")
    plt.ylabel(f"Количество домов")
    plt.show()
```

В данных по ценам - много выбросов, значений после 1 млн не много
В данных по количество спален также есть выбросы - домов с количеством спален больше 5, очень мало
Большинство домов имеют от 1 до 4 ванных домнат
Площадь большинства домов не превышает 4000 sqft
Большинство продающихся домов 1 или 2-х этажные, без подвала.
В значениях общей площади очень много выбросов, нет понимания об изменяемости параметра, также как и о годах постройки

Построим boxplot для каждого параметра

```python
plt.figure(figsize=(10,5))
for i in test_data_numerical:
    sns.boxplot(test_data[i], showfliers = False)
    plt.title(f'Метрика {i}')
    plt.xlabel(f"{i}")
    plt.ylabel(f"Количество домов")
    plt.show()
```

Уточним выводы:
Средняя цена дома - чуть более 400 000 у.е., максимальная стоимость приближается к 2 млн.
Среднее количество спален в доме 3, максимально 5
Большинство домов имеют по 2 ванные комнаты, редко 3.
Жилая площадь домов от 1500 до 2500 sqft, общая площадь домов от 5000 до 10000 sqft. Дома в 1 или 2 этажа, без подвала. Продаются дома постройки от 1950 по 2000 годов, ремонт производился в крайне малом количество домов.

Проанализаруем категориальные показатели

```python
plt.figure(figsize=(10,5))
for i in test_data_categorical:
    data = test_data[i].value_counts().reset_index()
    sns.barplot(y=data[i],x=data["index"])
    plt.title(f'Метрика {i}')
    plt.xlabel(f"{i}")
    plt.ylabel(f"Количество домов")
    plt.xticks(rotation=30)
    plt.show()
```

Дома по большей части не имеют вида на залив, содержатся в удовлетворительном состоянии.

Проанализируем влияния вещественных параметров на цены домов

```python
for i in test_data_numerical:
    sns.jointplot(x=test_data[i], y=test_data['price'], kind='reg')
    plt.title(f'Зависимость цены от метрики {i}')
    plt.xlabel(f"Значение метрики {i}")
    plt.ylabel(f"Цены домов")
    #plt.xticks(rotation=30)
    plt.show()
```

Выводы:
с увеличением количества спален, растет и цена дома. Однако, для домов с количеством спален больше ~5, стоимость начинает резко снижаться. Размер жилой площади напрямую влияет на стоимость дома, зависимость цены от общей площади не просматривается. Этажность также мало влияет на цену, также как и налииче подвала. Цена дома плавно увеличивается одновременно с уменьшением возраста дома. Цены на отремонтированные дома ожидаемо чуть выше.

Проанализируем влияния категориальных параметров на цены домов

```python
for i in test_data_categorical:
    sns.boxplot(x=test_data[i], y=test_data['price'], showfliers = False)
    plt.title(f'Зависимость цены от метрики {i}')
    plt.xlabel(f"Значение метрики {i}")
    plt.ylabel(f"Цены домов")
    #plt.xticks(rotation=30)
    plt.show()
```

Выводы:
вид на набережную значительно повышает стоимость дома. Оценка вида из окна слабо влияет на цену, однако для высоких оценок стоимость дома будет выше. Дома оценненных удовлетворительно и выше мало разнятся по средним стоимостям. Ожидаемо. с росто общей оценки дома, растет и его цена.

Поистроим матрицу корреляций по параметрам, которые в предыдущем анализе оказывали влияние на значение стоимости, и посмотрим на силу связей

```python
parameters = ["price","bedrooms","bathrooms","sqft_living","sqft_lot","floors","yr_built","yr_renovated","waterfront","view","condition","grade"]
price_corr = test_data[[*parameters]].corr()
plt.figure(figsize=(20,15))
sns.heatmap(price_corr, center=0, annot=True)
```

Выводы:
на цену сильно влияет размер жилой площади, а также общая оценка дома. Заметную роль в ценообразовании играет количество ванных комнат, количество спален практически не имеет влияния. Дома, имеющие вид на набережную хотя и стоят дороже, этот парамер не является ключевым при определении стоимости.

Изучим влияние параметров на цены домов с видом на набережную

```python
parameters = ["price","bedrooms","bathrooms","sqft_living","sqft_lot","floors","yr_built","yr_renovated","view","condition","grade"]
for i in [0,1]:
    data_waterfront = test_data[test_data["waterfront"]==i]
    price_corr = data_waterfront[[*parameters]].corr()
    plt.figure(figsize=(20,15))
    sns.heatmap(price_corr, center=0, annot=True)
    plt.title(f"Зависимости цен на дома {'с видом на набережную' if i==1 else 'без вида на набережную'}")
```

Для домов с видо на набережную самыми важными параметрами являются жилая площадь, количество ванных комнат и общая оценка. Количество спален также заметно увеличивает стоимость жилья. Умеренную роль играет год постройки
Для домов, не имеющих вида на набережную, количество спален слабо влияет на стоимость, год постройки практически не оказывает влияния на цену.
