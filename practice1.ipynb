## Задание2
<p>На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе, при помощи использования kwargs</p>

```python
goods = {
    'апельсины': 5,
    'яблоки': 4,
    'персики': 10,
    'киви': 1
}

def sum_fruits(**kwargs):
    try:
        return sum(kwargs.values())
    except:
        print(f"Ошибка во входящих данных")

print(sum_fruits(**goods))
```

### подзадача2
<p>Найдите выручку компании в зависимости от месяца</p>
<p>Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.</p>
<p>Используйте аннотирование типов.</p>

```python
def count_sales(dates: list, revenues:list) -> dict:
    months = list(set({i[:-3] for i in dates}))#список уникальных сочетаний год-месяц
    sales = list(zip(dates, revenues))#список кортежей из даты и суммы продаж
    month_sales = {month: 0 for month in months} 
    for key in month_sales.keys():
        for i in sales:            
            if key in i[0]:                
                month_sales[key] += i[1]
    return month_sales
```

```python
count_sales(dates, revenues)
```
