### Задача1

```python
house_data = pd.read_csv("kc_house_data.csv")
house_data.head()
```

```python
house_data["delta_renovated"] = house_data.apply(lambda x: 0 if x['yr_renovated'] == 0 else x['yr_renovated'] - x['yr_built'], axis=1)
#house_data["delta_renovated"] = [0 if x==0 else x-y for x,y in house_data[["yr_renovated","yr_built"]]]
house_data[(house_data.delta_renovated>0)]#проверили расчет
```

```python
house_data["date"] = house_data["date"].apply(lambda x: x.split("T")[0])#отбросили то, что должно было быть временем
house_data.head()
```

```python
house_data["year"] = house_data["date"].apply(lambda x: int(x[0:4]))
house_data["month"] = house_data["date"].apply(lambda x: int(x[4:6]))
house_data.head()                            
```

```python
house_data.drop(columns=["date", "zipcode", "lat", "long"], inplace=True)
house_data.info()
```

### Задача3

#### Cредние стоимости домов в зависимости от количества спален

```python
house_data.groupby("bedrooms", as_index = False)["price"].mean().round(1).sort_values("price",ascending = False)
```

#### Минимальная, средняя и максимальная стоимости домов в зависимости от состояния дома

```python
house_data.groupby("condition", as_index = False)["price"].agg({'Минимум':'min','Максимум':'max','Средняя':'mean'}).round(1)
```

#### Каких домов в зависимости от этажности и количества спален больше?
<p>Одно- и двухэтажных с количеством спален от 3х до 4х</p>

```python
house_data.groupby(["floors","bedrooms"], as_index = False)["id"].count().sort_values("id",ascending = False)
```

#### Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
house_data.pivot_table(index = "view", columns = "waterfront", values = "id", aggfunc = "count", fill_value = 0)
```

#### Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
med = house_data.pivot_table(index = "grade", columns = "condition", values = "id", aggfunc = "median" )
```
